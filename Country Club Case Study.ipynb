```python
import pandas as pd
import sqlite3 as sql
```

```python
database = 'country_club.db'
connection = sql.connect(database)
```

Q10: Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members!

```python
query = """WITH members AS (SELECT f.name AS facility, SUM(b.slots * f.membercost) AS member_rev FROM Bookings AS b LEFT JOIN Facilities AS f ON b.facid = f.facid WHERE b.memid != 0 GROUP BY facility),
guests AS (SELECT f.name AS facility, SUM(b.slots * f.guestcost) AS guest_rev FROM Bookings AS b LEFT JOIN Facilities AS f ON b.facid = f.facid WHERE b.memid = 0 GROUP BY facility)
SELECT m.facility, (member_rev+guest_rev) AS total_revenue
FROM members as m LEFT JOIN guests as g on m.facility = g.facility
WHERE total_revenue < 1000
ORDER BY total_revenue;"""
pd.read_sql_query(query, connection)
```

Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order

```python
query = """SELECT DISTINCT (m1.surname || ', ' || m1.firstname) AS name, (m2.surname || ', '|| m2.firstname) as recommended_by
FROM Members as m1 LEFT JOIN Members as m2 ON m1.recommendedby = m2.memid
ORDER BY m1.surname"""
pd.read_sql_query(query, connection)
```

Q12: Find the facilities with their usage by member, but not guests

```python
query = """SELECT f.name as facility, COUNT(*) AS member_usage 
FROM Bookings as b
LEFT JOIN Facilities as f
USING(facid)
WHERE memid != 0
GROUP BY facility""" 
pd.read_sql_query(query, connection)
```

Q13: Find the facilities usage by month, but not guests

```python
query = """SELECT f.name as facility, strftime('%m', starttime) AS month, COUNT(*) AS member_usage 
FROM Bookings as b
LEFT JOIN Facilities as f
USING(facid)
WHERE memid != 0
GROUP BY facility, month"""

pd.read_sql_query(query, connection)
```
